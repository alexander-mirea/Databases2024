# Описание таблиц

```python
%%sql
SELECT * FROM sqlite_master; -- выводит все объекты БД
```

## Таблица employees
Таблица хранит данные сотрудников, такие как _employee id_, _last name_, _fist name_, и т.д. Также есть поле _ReportsTo_, которое показывает, кто чей подчиненный

```python
%%sql
SELECT * FROM Employees;
```

## Таблица customers
Таблица хранит данные по клиентам. Поле _SupportRepid_ является внешним ключом к таблице _employees_ и обозначает, какой сотрудник закреплен за клиентом

```python
%%sql
SELECT *
FROM Customers LIMIT 5;
```

## Таблицы invoices и invoice_items
Эти две таблицы хранят информацию по заказам. Таблица _Invoice_ хранит заголовок заказа, а также внешний ключ к таблице _customers_. Таблица _invoice_items_ хранит детали заказа, а также внешний ключ к таблицам _invoices_, _tracks_

```python
%%sql
SELECT *
FROM invoices LIMIT 10;
```

```python
%%sql
SELECT *
FROM invoice_items LIMIT 10;
```

## Таблица artists
Таблица хранит данные об артистах (только id и имя)

```python
%%sql
SELECT *
FROM artists LIMIT 10;
```

## Таблица albums
Хранит информацию об альбомах, в ней есть 2 внешних ключа на таблицы _artists_ и _tracks_

```python
%%sql
SELECT *
FROM albums LIMIT 10;
```

## Таблица media_types
Хранит информацию о всевозможных форматах

```python
%%sql
SELECT *
FROM media_types LIMIT 10;
```

## Таблица genres
Хранит информацию о всевозможных жанрах

```python
%%sql
SELECT *
FROM genres LIMIT 10;
```

## Таблица tracks
Хранит информацию о композициях. В данной таблице есть внешние ключи на _albums_, _mediatypes_ и _genres_

```python
%%sql
SELECT *
FROM tracks LIMIT 10;
```

## Таблицы playlists и playlist_track
Таблица _playlists_ хранит список плейлистов. Таблица _playlist_track_ - соединительная таблица в связи многие-ко-многим (_playlists_ и _tracks_)

```python
%%sql
SELECT *
FROM playlists LIMIT 10;
```

```python
%%sql
SELECT *
FROM playlist_track LIMIT 10 OFFSET 210;
```
